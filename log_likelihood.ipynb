```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 5, 1
x = np.random.normal(mu, sigma, 10)
```

```python
prob = (1/(np.sqrt(2*np.pi)*sigma))*np.exp(-((x-mu)**2/2*(sigma**2)))
likelihood = np.prod(prob)
print(likelihood)
```

```python
a = []
mean = np.linspace(0,10,50)
prob = np.zeros(len(mean)).reshape(50,1)
for i,mu in enumerate(mean):
    prob[i,0] = np.prod((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-((x-mu)**2/2*(sigma**2))))
    log_prob = (-0.5*np.sum(x-mu)**2) - 0.5*np.log(1) - 0.5*np.log(2*np.pi)
    a.append(log_prob)
    likelihood = np.prod(prob)
```

```python
plt.plot(mean,a)
```
