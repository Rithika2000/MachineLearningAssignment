```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
import seaborn as sns

def dataset():
    x=np.linspace(-1,1,10)
    y=0.5*x-0.3+np.random.normal(0,0.2,10)
    return y
sample_pts=dataset()


likelihood_grid=norm(0,0.2).pdf(sample_pts)
prior_grid=norm(0,0.71).pdf(sample_pts)
likelihood_grid,prior_grid=np.meshgrid(likelihood_grid,prior_grid)

def P(J,K):
    return np.multiply(J,K)

g=P(likelihood_grid,prior_grid)
plt.figure(figsize=(12,9))
ap=sns.heatmap(g, xticklabels=True, yticklabels=True,center=0,annot=True,linewidth=0.2,cmap="BuPu")
ap.invert_yaxis()
plt.show()
```
